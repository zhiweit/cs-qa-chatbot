```python
import pandas as pd
# import dataset1 (Github repository by Kunal Kumar)
questions = []
answers = []
with open('final1.txt', 'r') as file:
    lines = file.readlines()

    for i in range(0, len(lines), 2):
        question = lines[i].strip()  
        if i + 1 < len(lines):  
            answer = lines[i + 1].strip()
        else:
            answer = ''  

        questions.append(question)
        answers.append(answer)

df1 = pd.DataFrame({'Question': questions, 'Answer': answers})
```

```python
# import dataset2 (Hugging Face - data-science-qa): 

import pandas as pd

df2 = pd.read_csv("hf://datasets/team-bay/data-science-qa/DataScienceDataset.csv")
df2 = df2[['question', 'answer']]
```

```python
# import dataset3 (Kaggle - Software Engineering Interview Questions Dataset)

df3 = pd.read_csv('SoftwareQuestions.csv', encoding='ISO-8859-1')

df3 = df3[['Question', 'Answer']]
```

```python
# import dataset4 (Kaggle - Computer Science Theory QA Dataset)

import json
import pandas as pd

with open('intents.json', 'r') as file:
    data = json.load(file)

questions = []
responses = []

for intent in data['intents']:
    patterns = intent['patterns']
    response_list = intent['responses']
    
    for pattern in patterns:
        questions.append(pattern)
        responses.append(response_list[0])

df4 = pd.DataFrame({'Question': questions, 'Answer': responses})
```

```python
df4
```

```python
questions_combined = pd.concat(
    [df1['Question'], df2['question'], df3['Question'], df4['Question']],
    axis=0
).reset_index(drop=True)

answers_combined = pd.concat(
    [df1['Answer'], df2['answer'], df3['Answer'], df4['Answer']],
    axis=0
).reset_index(drop=True)

combined_df = pd.DataFrame({
    'Question': questions_combined,
    'Answer': answers_combined
})

combined_df
```

```python
# Check for NULL values
combined_df.isna().sum()
```

```python
import pandas as pd
import re
from bs4 import BeautifulSoup

# Function to clean text
def clean_text(text):
    # Remove HTML tags
    text = BeautifulSoup(text, "html.parser").get_text()
    # Remove special characters and unnecessary punctuation
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# Apply the cleaning function to the DataFrame
combined_df['Question'] = combined_df['Question'].apply(clean_text).str.lower()
combined_df['Answer'] = combined_df['Answer'].apply(clean_text).str.lower()
combined_df
```

```python
# Save the DataFrame to a CSV file
combined_df.to_csv('combined_data.csv', index=False)
```
